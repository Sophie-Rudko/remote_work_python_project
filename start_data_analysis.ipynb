```python
def transform_stress(x):
    if x == "Low":
        return 1
    if x == "Medium":
        return 2
    if x == "High":
        return 3
df["Stress_Level_Rating"] = df.Stress_Level.apply(transform_stress)
```

```python
def transform_satisfaction(x):
    if x == "Unsatisfied":
        return 1
    if x == "Neutral":
        return 2
    if x == "Satisfied":
        return 3
df["Satisfaction_with_Remote_Work_Rating"] = df.Satisfaction_with_Remote_Work.apply(transform_satisfaction)
```

```python
def transform_sleep(x):
    if x == "Poor":
        return 1
    if x == "Average":
        return 2
    if x == "Good":
        return 3
df["Sleep_Quality_Rating"] = df.Sleep_Quality.apply(transform_sleep)
```

```python
import numpy as np
```

```python
ratings_for_regions = dict()
for region in df.Region.unique():
    stress = round(df[df.Region == region].Stress_Level_Rating.mean(), 2)
    satisfaction = round(df[df.Region == region].Satisfaction_with_Remote_Work_Rating.mean(), 2)
    sleep = round(df[df.Region == region].Sleep_Quality_Rating.mean(), 2)
    support = round(df[df.Region == region].Company_Support_for_Remote_Work.mean(), 2)
    isolation = round(df[df.Region == region].Social_Isolation_Rating.mean(), 2)
    balance = round(df[df.Region == region].Work_Life_Balance_Rating.mean(), 2)
    ratings_for_regions[region] = [stress, satisfaction, sleep, support, isolation, balance]
print(len(list(ratings_for_regions.values())), len(df.Region.unique()))

df_regs = pd.DataFrame(np.array(list(ratings_for_regions.values())), columns=df.Region.unique())
df_regs
```

```python
fig, axs = plt.subplots(3, 2, figsize=(15, 12))
labels=["Stress Level", "Satisfaction with Remote Work", "Sleep Quality Rating", "Company Support for Remote Work", "Social Isolation Rating", "Work Life Balance Rating"]
colors = ['chocolate', 'lightcoral', 'indianred', 'tomato', 'brown', 'darksalmon', 'sandybrown']
x = 0
y = 0
for region in df.Region.unique():
    bars = axs[x % 3, y % 2].barh(labels, df_regs[region], color=colors[y], height=0.5)
    axs[x % 3, y % 2].set_title(region)  
    axs[x % 3, y % 2].set_xlabel('Rating')  
    axs[x % 3, y % 2].set_ylabel('Factor')  
    offset = 0.2
    for bar in bars:
        axs[x % 3, y % 2].text(bar.get_width() + 0.05,  # Смещение за пределы графика
                               bar.get_y() + bar.get_height() / 2,  # По центру бара
                               f'{bar.get_width():.2f}',  # Значение с двумя знаками после запятой
                               va='center', ha='left', fontsize=10)
    axs[x % 3, y % 2].set_xlim(0, max(df_regs[region]) * 1.12)
    
    y += 1
    if y % 2 != 0:
        x += 1

plt.tight_layout()  
plt.show()
```

```python
labels = ["Stress Level", "Satisfaction with Remote Work", "Sleep Quality Rating",
          "Company Support for Remote Work", "Social Isolation Rating", 
          "Work Life Balance Rating"]
colors = ['chocolate', 'lightcoral', 'indianred', 'tomato', 'brown', 'darksalmon', 'sandybrown']

regions = df["Region"].unique()
x = np.arange(len(labels))  # Positions for the groups
width = 0.8 / len(regions)  # Adjust bar width based on the number of regions

# Create the plot
fig, ax = plt.subplots(figsize=(15, 8))

# Plot bars for each region
for i, region in enumerate(regions):
    region_data = df_regs[region]
    ax.bar(
        x + i * width, 
        region_data, 
        width, 
        label=region, 
        color=colors[i % len(colors)]
    )

# Add labels, title, and legend
ax.set_xticks(x + width * (len(regions) - 1) / 2)
ax.set_xticklabels(labels, rotation=45, ha="right")
ax.set_title("Comparison of Ratings Across Regions")
ax.set_xlabel("Factors")
ax.set_ylabel("Average Rating")
ax.legend(title="Regions", bbox_to_anchor=(1.05, 1), loc='upper left')

# Add value annotations
for i, region in enumerate(regions):
    region_data = df_regs[region]
    for j, value in enumerate(region_data):
        ax.text(
            x[j] + i * width, 
            value + 0.05, 
            f"{value:.2f}", 
            ha="center", va="bottom", fontsize=8
        )

plt.tight_layout()
plt.show()
```
